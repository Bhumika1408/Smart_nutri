```python
demographics = pd.read_csv('NHANES Demographics.csv')
diabetes = pd.read_csv('NHANES Diabetes Questionnaire.csv')
blood_pressure = pd.read_csv('NHANES Blood Pressure Quesstionnaire.csv')
nutrients_day1 = pd.read_csv('NHANES Total Nutrients Day 1.csv')
nutrients_day2 = pd.read_csv('NHANES Total Nutrients Day 2.csv')
```

```python
# Merge health metrics
health_data = pd.merge(diabetes, blood_pressure, on='SEQN', how='inner')
health_data = pd.merge(health_data, demographics, on='SEQN', how='inner')
```

```python
print(diabetes.columns)
```

```python
print(blood_pressure.columns)
```

```python
print("Diabetes SEQN Data Type:", diabetes['SEQN'].dtype)
print("Blood Pressure SEQN Data Type:", blood_pressure['SEQN'].dtype)
```

```python
# Convert SEQN to int in both datasets
diabetes['SEQN'] = diabetes['SEQN'].astype(int)
blood_pressure['SEQN'] = blood_pressure['SEQN'].astype(int)
```

```python
# Merge datasets
health_data = pd.merge(diabetes, blood_pressure, on='SEQN', how='inner')

print("Merged Data:")
print(health_data.head())
```

```python
print("Unique SEQN in Diabetes:", diabetes['SEQN'].unique())
print("Unique SEQN in Blood Pressure:", blood_pressure['SEQN'].unique())
```

```python
merged_ids = set(health_data['SEQN'])
diabetes_ids = set(diabetes['SEQN'])
blood_pressure_ids = set(blood_pressure['SEQN'])

missing_in_bp = diabetes_ids - merged_ids
missing_in_diabetes = blood_pressure_ids - merged_ids

print("Missing in Blood Pressure:", missing_in_bp)
print("Missing in Diabetes:", missing_in_diabetes)
```
